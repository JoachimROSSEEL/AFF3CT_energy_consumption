# Plot FER and BER according to polar code decoders

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

# Code parameters

```python
# Codeword size
N = 4096 # 32768 #4096

# Number of information bits
K = 3072 # 16384 # 2048 # 1024 # 3072

# CRC 
CRC = "0x8005" # "0x04C11DB7" # "0x8005"
```

# Files reading

```python
# FER 
files_FER_path = "FER/"
# _nodes_{R0,R1,R0L,REP,REPL,SPC}
file_SC = np.loadtxt("FER/FER_polar_" + str(N) + "_" + str(K) +"_CRC_" + CRC + "_Decoder_polar_SC_fast_sys.txt")
file_SCL = np.loadtxt("FER/FER_polar_" + str(N) + "_" + str(K) +"_CRC_" + CRC + "_Decoder_polar_SCL_fast_CA_sys_L_4.txt")
# file_SC = np.loadtxt("FER/FER_polar_4096_2048_CRC_0x8005_Decoder_polar_SC_fast_sys.txt")
# file_SCL = np.loadtxt("FER/FER_polar_4096_2048_CRC_0x8005_Decoder_polar_SCL_fast_CA_sys_L_4.txt")
# file_ASCL = np.loadtxt("FER/FER_polar_64_32_CRC_0x3_Decoder_polar_ASCL_fast_CA_sys_L_4.txt")
```

# Plotting FER

```python
width = 7
height = 5 
ms = 8
lw = 3
label_size = 12

fig_name = "FER"

fig_fer, (ax_fer) = plt.subplots(num=fig_name, figsize=(width, height))

ax_fer.plot(file_SC[:,0], file_SC[:,1], ms=ms, linewidth=lw, marker='o', color='green')
ax_fer.plot(file_SCL[:,0], file_SCL[:,1], ms=ms, linewidth=lw, marker='o', color='red')
# ax_fer.plot(file_ASCL[:,0], file_ASCL[:,1], ms=ms, linewidth=lw, marker='o', color='blue')
ax_fer.set_xlim((2.9, 3.1))
# ax_fer.set_ylim((1e-4, 2))
ax_fer.set_xlabel("SNR(dB)", fontsize=label_size)
ax_fer.set_ylabel("FER", fontsize=label_size)
ax_fer.set_yscale("log")
ax_fer.grid(True, which='both', axis='both')
ax_fer.tick_params(axis='both', which='major', labelsize=label_size)  
ax_fer.tick_params(axis='both', which='minor', labelsize=label_size)
```
